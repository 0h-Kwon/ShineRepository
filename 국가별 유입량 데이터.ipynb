## 국내항공사 운항 공항

```python
file_dir = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow'
save_file = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/airport_list(domestic).csv'
all_file = glob.glob(os.path.join(file_dir, '인천(국내)_20*'))

airport_data = []
for file in all_file:
    airport = pd.read_excel(file)
    airport = airport.drop([0,1,2], axis=0)
    airport = airport.drop(airport.columns[0], axis = 1)
    airport = airport.rename({'Unnamed: 1':'Airport',
                              'Unnamed: 2':'A/P',
                              'Unnamed: 3':'PAX'}, axis = 1)
    airport_data.append(airport)

dataCombine = pd.concat(airport_data, axis = 0, ignore_index = True)
dataCombine = dataCombine.drop(dataCombine.columns[[3,4]], axis=1)
dataCombine.to_csv(save_file, encoding = 'utf-8-sig')

airport_list = dataCombine
airport_list = airport_list.groupby(['Airport']).count()
airport_list = airport_list.drop(['A/P','PAX'], axis=1)
airport_list.drop(['(RPVM)','KAPG (KAPG)','KAVQ (KAVQ)','사천 (RKPS)','인천 (RKSI)',
                   '김포 (RKSS)','김해 (RKPK)','제주 (RKPC)','청주 (RKTU)','인천* (RKZZ)',
                   '대구 (RKTN)','광주 (RKJJ)','울산 (RKPU)'], inplace = True)

airport_list.to_csv('C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/domestic_airport.csv', encoding = 'utf-8-sig')
airport_list
```

## 외국항공사 운항 공항

```python
# csv파일들이 있는 위치
file_dir = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow'
# 병합하고 저장하려는 파일명
save_file = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/airport_list(overseas).csv'
# 해당 이름을 가진 파일들 모으기(glob()함수)
all_file = glob.glob(os.path.join(file_dir, '인천(외국)_20*'))

# 읽어 들인 csv파일 내용을 저장할 리스트
airport_data = []
# 반복문으로 csv 파일 읽어오기 및 내용 리스트에 추가
for file in all_file:
    airport = pd.read_csv(file, engine = 'python')
    airport = airport.drop([0], axis=0)
    airport.columns = ['Airport','A/P','PAX']
    airport_data.append(airport)
# 리스트 내용 병합(아래로 병합, 인덱스 무시), csv파일로 저장
dataCombine = pd.concat(airport_data, axis = 0, ignore_index = True)
dataCombine.to_csv(save_file, encoding = 'utf-8-sig')

airport_list = dataCombine
# 같은 공항끼리 묶기
airport_list = airport_list.groupby(['Airport']).count()
# 불필요한 열, 행 삭제
airport_list = airport_list.drop(['A/P','PAX'], axis=1)
airport_list.drop(['(WIPP)','FJDG (FJDG)','VIPK (VIPK)','ZSNT (ZSNT)'], inplace = True)
# 저장
airport_list.to_csv('C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/overseas_airport.csv', encoding = 'utf-8-sig')
airport_list
```

# 일일 유입 항공, 승객 수 데이터 프레임 생성

## 국내항공사 유입 데이터

```python
# csv파일들이 있는 위치
file_dir = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow'
# 병합하고 저장하려는 파일명
save_file = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/domestic_airline.csv'
# 해당 이름을 가진 파일들 모으기(glob()함수)
inflow_DA = glob.glob(os.path.join(file_dir, '인천(국내)_*'))
# 읽어 들인 csv파일 내용을 저장할 리스트
DA_data = []
# 반복문으로 csv 파일 읽어오기 및 내용 리스트에 추가
for file in inflow_DA:
    domestic = pd.read_excel(file)
    domestic = domestic.drop([0,1,2], axis = 0)
    domestic = domestic.drop(domestic.columns[0], axis = 1)
    domestic = domestic.rename({'Unnamed: 1':'Airport','Unnamed: 2':'A/P','Unnamed: 3':'PAX'}, axis = 1)
    # 같은 공항 값끼리 sum
    domestic = domestic.groupby(['Airport']).sum()
    DA_data.append(domestic)

# 리스트 내용 병합(아래로 병합, 인덱스 무시), csv파일로 저장
dataCombine = pd.concat(DA_data, axis = 1, ignore_index = True)
# 결측값 채우기
dataCombine = dataCombine.fillna(0)
# 불필요한 행 제거
dataCombine.drop(['(RPVM)','KAPG (KAPG)','KAVQ (KAVQ)','사천 (RKPS)','인천 (RKSI)',
                  '김포 (RKSS)','김해 (RKPK)','제주 (RKPC)','청주 (RKTU)','인천* (RKZZ)',
                  '대구 (RKTN)','광주 (RKJJ)','울산 (RKPU)'], inplace = True)
# 저장저장~~
dataCombine.to_csv(save_file, encoding = 'utf-8-sig')
dataCombine
```

## 외국항공사 유입 데이터

```python
# csv파일들이 있는 위치
file_dir = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow'
# 병합하고 저장하려는 파일명
save_file = r'C:/Users/shine/Desktop/BigData/Class/project/SecondPjt/data/overseas inflow/overseas_airline.csv'
# 해당 이름을 가진 파일들 모으기(glob()함수)
inflow_OA = glob.glob(os.path.join(file_dir, '인천(외국)_*'))
# 읽어 들인 csv파일 내용을 저장할 리스트
OA_data = []
# 반복문으로 csv 파일 읽어오기 및 내용 리스트에 추가
for file in inflow_OA:
    overseas = pd.read_csv(file, engine = 'python')
    overseas = overseas.drop([0], axis=0)
    overseas.columns = ['Airport','A/P','PAX']
    overseas = overseas.astype({'A/P' : float, 'PAX' : float})
    overseas = overseas.groupby(['Airport']).sum()
    OA_data.append(overseas)
    
# 리스트 내용 병합(아래로 병합, 인덱스 무시), csv파일로 저장
dataCombine = pd.concat(OA_data, axis = 1, ignore_index = True)
dataCombine.drop(['(WIPP)','FJDG (FJDG)','VIPK (VIPK)','ZSNT (ZSNT)'], inplace = True)
# 결측값 채우기 및 정수형으로 데이터 타입 변경
dataCombine = dataCombine.fillna(0)
dataCombine = dataCombine.astype(int)
# 저장저장~~
dataCombine.to_csv(save_file, encoding = 'utf-8-sig')
dataCombine
```
