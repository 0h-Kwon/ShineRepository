```
import pandas as pd
import tensorflow as tf
```

```
iris_data = pd.read_csv('https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv')
print(iris_data.columns)
iris_data.head(3)
```

```
iris = pd.get_dummies(iris_data)   # onehot-encoding : 범주형 데이터를 0, 1로 바꿔주는 것
iris.tail(3)
```

```
x_data = iris[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
y_data = iris[['품종_setosa',	'품종_versicolor',	'품종_virginica']]
print(x_data.shape, y_data.shape)
```

```
X = tf.keras.layers.Input(shape=[4])
Y = tf.keras.layers.Dense(3, activation='softmax')(X)   # softmax : 확률로 예측하기 위해 사용(0~100%)
                                                        # activation : 활성화 함수
                                                        # softmax와 sigmoid의 차이는 무엇인가
model = tf.keras.models.Model(X, Y)
model.compile(loss='categorical_crossentropy',          # 분류에 사용하는 loss는 'crossentropy', 회귀는 'mse'
              metrics='accuracy')
```

```
model.fit(x_data, y_data, epochs=100)
```

```
print('Predictions: ', model.predict(x_data[0:5]))
```
