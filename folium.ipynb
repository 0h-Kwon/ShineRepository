```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import folium
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
report = pd.read_csv('report.csv')
report
```

```python
report = report.groupby('시군구명')['총생활인구수'].mean().to_frame().reset_index()
report = report.drop(report[report['시군구명']=='서울시'].index)
report
```

```python
report = report.astype('int', errors='ignore')
report
```

```python
gj_path = 'TL_SCCO_SIG.geojson'
gj_str = json.load(open(gj_path, encoding='utf-8'))
gj_str
```

```python
map = folium.Map(location=[37.5502, 126.982], zoom_start=11)
map.choropleth(geo_data = gj_str,
               data = report,
               columns = ('시군구명','총생활인구수'),
               key_on = 'feature.properties.SIG_KOR_NM',
               fill_color = 'PuRd',
               legend_name = 'Rate(%)')
map
```
