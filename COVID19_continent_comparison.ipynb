```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid19 = pd.read_csv('COVID19_global_data.csv')
```

```python
covid19
```

```python
# 불필요한 columns 삭제 : 컬럼명에 있는 사소한 부분(띄어쓰기 등 에러 발생요인이 됨) 주의.
covid19 = covid19.drop([' WHO_region', ' Country_code'], axis=1)
```

```python
# 컬럼명 변경
covid19 = covid19.rename(columns={'Date_reported':'date',' Country':'country',' New_cases':'n_cases',' Cumulative_cases':'c_cases',' New_deaths':'n_deaths', ' Cumulative_deaths':'c_deaths'})
covid19
```

```python
# 국가별 데이터 프레임 생성 후 left join으로 합치기
# nc(new cases), cc(cumulative cases), nd(new deaths), cd(cumulative deaths)
# East Asia
korea = covid19[covid19.country == 'Republic of Korea']
china = covid19[covid19.country == 'China']
japan = covid19[covid19.country == 'Japan']
e_asia = china.merge(japan, how = "left", on = 'date').merge(korea, how = "left", on = 'date')
e_asia = e_asia.rename(columns={'n_cases_x':'chn_nc','c_cases_x':'chn_cc','n_deaths_x':'chn_nd','c_deaths_x':'chn_cd',
                                'n_cases_y':'jpn_nc','c_cases_y':'jpn_cc','n_deaths_y':'jpn_nd','c_deaths_y':'jpn_cd',
                                'n_cases':'kor_nc','c_cases':'kor_cc','n_deaths':'kor_nd','c_deaths':'kor_cd'})
e_asia.to_csv("e_asia.csv")
e_asia
```
