# 서울 일일 확진자

```python
seoulData = pd.read_csv('C:/Users/shine/Desktop/BigData/Mentoring/관련데이터/시도별 코로나 일일 확진자 현황(서울).csv', engine='python' )
seoulData
```

```python
seoul = seoulData[['확진일']]
seoul['확진자 수'] = 1
seoul
```

```python
date = seoul['확진일']
```

```python
def make_date(date):
    if date.find('-') == 1:
        date = '2020-'+'0'+date
    else:
        date = '2020-'+date
    if len(date) < 10:
        date = date[:8]+'0'+date[-1]
    return date        
date
```

```python
date = seoul['확진일'].str.rstrip('.').str.replace('.','-').apply(make_date)
```

```python
date = pd.to_datetime(date, format='%Y-%m-%d')
seoul['확진일'] = date
seoul
```

```python
seoul = seoul.groupby('확진일').count()
seoul
```

# 전국&시도 일일 확진자
